## Popularity Based Recommender System

```python
ratings.columns
```

```python
ratings_with_title = ratings.merge(books, on='ISBN')
```

```python
ratings_with_title.info()
```

```python
number_of_rating_df = ratings_with_title.groupby('Book-Title')["Book-Rating"].count().reset_index()
number_of_rating_df.rename(columns={'Book-Rating': 'num_ratings'}, inplace=True)
number_of_rating_df.head()
```

```python
avg_rating_df = ratings_with_title.groupby('Book-Title')["Book-Rating"].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_rating_df.head()
```

```python
popular_books = number_of_rating_df.merge(avg_rating_df, on='Book-Title')
popular_books.head()
```

```python
popular_top_25_df = popular_books[popular_books['num_ratings'] > 200].sort_values(by='avg_rating', ascending=False).head(25)
```

```python
popular_top_25_df
```

```python
popular_top_25_df.merge(books, on='Book-Title').shape
```

```python
popular_top_25_book_df = popular_top_25_df.merge(books, on='Book-Title').drop_duplicates(subset='Book-Title')[['Book-Title', 'ISBN', 'Book-Author',
       'Year-Of-Publication', 'Publisher', 'Image-URL-S', 'Image-URL-M',
       'Image-URL-L']]
popular_top_25_book_df.columns
```

## Collaborative Filtering Based Recommender System

```python
# ratings_with_title = ratings_with_title.merge(books, on='ISBN')
ratings_with_title.info()
```

```python
# Number of users who have rated the books
ratings_with_title.groupby('User-ID')['Book-Rating'].count().sort_values(ascending=False)
```

```python
# filter the users who have rated more than 200 books
x = ratings_with_title.groupby('User-ID')['Book-Rating'].count() > 200
x = x[x].index
filtered_user_ratings_df = ratings_with_title[ratings_with_title['User-ID'].isin(x)]
filtered_user_ratings_df.info()
```

```python
y = filtered_user_ratings_df.groupby("Book-Title").count()["Book-Rating"] >= 50
famous_books = y[y].index
final_ratings = filtered_user_ratings_df[filtered_user_ratings_df['Book-Title'].isin(famous_books)]
```

```python
final_ratings.info()
```

```python
final_ratings.duplicated().sum()
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt
```

```python
pt.fillna(0, inplace=True)
pt
```

```python
pt[254].value_counts()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(pt)
```

```python
cosine_similarity(pt).shape
```

```python
def recommend(book_name):
	index = pt.index.get_loc(book_name)
	similarity_scores = list(enumerate(cosine_similarity(pt)[index]))
	similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
	similarity_scores = similarity_scores[1:6]
	book_indices = [i[0] for i in similarity_scores]
	recommended_books = pt.index[book_indices]
	return recommended_books
```

```python
recommend('1984')
```

```python
def recommend(book_name):
	index = np.where(pt.index == book_name)[0][0]
	similar_items = sorted(list(enumerate(cosine_similarity(pt)[index])), key=lambda x: x[1], reverse=True)[1:6]
	data = []
	for i in similar_items:
		item = []
		temp_df = books[books['Book-Title'] == pt.index[i[0]]]
		item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
		item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
		item.extend(list(temp_df.drop_duplicates('Book-Title')['Year-Of-Publication'].values))
		item.extend(list(temp_df.drop_duplicates('Book-Title')['Publisher'].values))
		item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))
		
		data.append(item)
	return data
```
